```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
import torch
from torchvision.utils import make_grid
```

```python
hdf5_path = '/home/haselab/projects/sakai/Adversarial-Perception-aware-Perturbation/output/pets_ddim_n50_p20_s3.5.hdf5'
with h5py.File(hdf5_path, 'r') as hdf:
    # 
    classes = list(hdf.keys())
    print(f"Classes: {classes}")

    for cls in classes:
        # 
        images = list(hdf[cls].keys())
        print(f"Images in {cls}: {images}")

        for img_name in images:
            # 
            dataset = hdf[f"{cls}/{img_name}"]

            # 
            img = np.array(dataset)

            # 
            label = dataset.attrs.get('label', None)
            print(f"Loaded image {img_name} from class {cls} with label {label}")

            # 
            plt.imshow(img)
            plt.title(f"Class: {cls}, Label: {label}")
            plt.show()
```
